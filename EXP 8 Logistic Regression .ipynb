```python
data=pd.read_csv("heart.csv")
```

```python
x=data.drop("target", axis=1)
y=data["target"]
```

```python
#splitting the data into training and testing data sets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2 ,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
log.fit(x_train, y_train)
```

```python
y_pred = log.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score 
```

```python


accuracy_score (y_test,y_pred)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
labels = np.unique(y_test)  # Get unique class labels
cm = pd.DataFrame(cm, index=labels, columns=labels)
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Reds', linewidths=1, linecolor='black')  

plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Select two features and target
X_vis = x_train[['age', 'chol']]
y_vis = y_train

# Train Logistic Regression
model = LogisticRegression()
model.fit(X_vis, y_vis)

# Create grid
xx, yy = np.meshgrid(np.linspace(X_vis['age'].min()-1, X_vis['age'].max()+1, 500),
                     np.linspace(X_vis['chol'].min()-1, X_vis['chol'].max()+1, 500))

# Predict probabilities for each point in grid
Z = model.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1]
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8,6))

# Plot decision boundary (probability=0.5 contour)
plt.contour(xx, yy, Z, levels=[0.5], colors='k', linewidths=2)

# Plot data points
plt.scatter(X_vis['age'], X_vis['chol'], c=y_vis, cmap='coolwarm', edgecolor='k', s=70)

plt.title("Logistic Regression Decision Boundary: Age vs Cholesterol")
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.grid(True)
plt.show()
```

###  Conclusion:
The logistic regression model effectively estimates the probability of a binary outcome by fitting the data to a logistic function, enabling accurate classification and prediction.

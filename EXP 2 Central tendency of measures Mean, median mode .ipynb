```python
import numpy as np
import pandas as pd
```

```python
data = [65, 70, 75, 80, 85, 90, 75, 85, 95, 100]
```

```python
arr = np.array(data)
print("Data as NumPy Array:", arr)
```

```python
mean_val = np.mean(arr)
print("Mean:", mean_val)
```

```python
median_val = np.median(arr)
print("Median:", median_val)
```

```python
series = pd.Series(arr)
mode_val = series.mode()
print("Mode:", mode_val.tolist())
```

```python
df = pd.DataFrame({'Marks': data})
print("Mean (Pandas):", df['Marks'].mean())
print("Median (Pandas):", df['Marks'].median())
print("Mode (Pandas):", df['Marks'].mode().tolist())
```

```python
import matplotlib.pyplot as plt

plt.hist(data, bins=5, color='lightblue', edgecolor='black')
plt.axvline(mean_val, color='red', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(median_val, color='green', linestyle='dashed', linewidth=2, label='Median')
for m in mode_val:
    plt.axvline(m, color='purple', linestyle='dotted', linewidth=2, label='Mode')
plt.legend()
plt.title('Distribution with Mean, Median, and Mode')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
data=pd.read_csv("heart.csv")
```

```python
x=data.drop("target", axis=1)
y=data["target"]
```

```python
#splitting the data into training and testing data sets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2 ,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score 
```

```python
knn=KNeighborsClassifier()
```

```python
knn.fit(x_train, y_train)
```

```python
y_pred=knn.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
```

```python
accuracy
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.patches as mpatches

# Select two features and target
X_vis = x_train[['age', 'chol']]
y_vis = y_train

# Train KNN
knn = KNeighborsClassifier(n_neighbors=7)  # try adjusting k
knn.fit(X_vis, y_vis)

# Create grid for decision boundary
xx, yy = np.meshgrid(np.linspace(X_vis['age'].min()-1, X_vis['age'].max()+1, 500),
                     np.linspace(X_vis['chol'].min()-1, X_vis['chol'].max()+1, 500))

Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8,6))

# Plot decision regions
plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')

# Plot data points with edge color and bigger size
scatter = plt.scatter(X_vis['age'], X_vis['chol'], c=y_vis, cmap='coolwarm', edgecolor='k', s=70)

plt.title("KNN Decision Boundary: Age vs Cholesterol")
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.grid(True)

# Add legend manually for clarity
classes = np.unique(y_vis)
colors = ['#FF6F61', '#61A0FF']  # Match coolwarm colors roughly
patches = [mpatches.Patch(color=colors[i], label=f'Class {classes[i]}') for i in range(len(classes))]
plt.legend(handles=patches)

plt.show()
```

###  Conclusion:
The KNN algorithm successfully classifies data points by assigning them the class most common among their nearest neighbors, demonstrating effective performance in pattern recognition and classification tasks.
